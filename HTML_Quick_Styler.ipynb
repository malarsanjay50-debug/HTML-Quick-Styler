```python
import gradio as gr
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
model_name = "ibm-granite/granite-3.3-2b-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)

def generate_ai_text(prompt):

    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

    output = model.generate(
        **inputs,
        max_new_tokens=300,
        temperature=0.7,
        do_sample=True
    )

    return tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
class ColorPaletteGenerator:

    palettes = {

        "modern": {
            "primary": "#6366F1",
            "secondary": "#8B5CF6",
            "accent": "#F59E0B",
            "background": "#0F172A",
            "surface": "#1E293B",
            "text": "#F1F5F9"
        },

        "ocean": {
            "primary": "#0284C7",
            "secondary": "#06B6D4",
            "accent": "#22D3EE",
            "background": "#082F49",
            "surface": "#0C4A6E",
            "text": "#E0F2FE"
        },

        "forest": {
            "primary": "#16A34A",
            "secondary": "#22C55E",
            "accent": "#A3E635",
            "background": "#052E16",
            "surface": "#14532D",
            "text": "#DCFCE7"
        },

        "sunset": {
            "primary": "#F97316",
            "secondary": "#FB7185",
            "accent": "#FACC15",
            "background": "#431407",
            "surface": "#7C2D12",
            "text": "#FFF7ED"
        }

    }

    def get_palette(self, style):

        return self.palettes.get(style, self.palettes["modern"])
```

```python
def generate_css(colors):

    css = f"""
    :root {{
        --primary:{colors['primary']};
        --secondary:{colors['secondary']};
        --accent:{colors['accent']};
        --bg:{colors['background']};
        --surface:{colors['surface']};
        --text:{colors['text']};
    }}

    body {{
        margin:0;
        font-family:Arial, sans-serif;
        background:var(--bg);
        color:var(--text);
    }}

    header {{
        background:var(--surface);
        padding:20px;
        position:sticky;
        top:0;
    }}

    .hero {{
        padding:100px;
        text-align:center;
        background:linear-gradient(135deg,var(--primary),var(--secondary));
    }}

    .features {{
        display:grid;
        grid-template-columns:repeat(auto-fit,minmax(200px,1fr));
        gap:20px;
        padding:40px;
    }}

    .card {{
        background:var(--surface);
        padding:20px;
        border-radius:10px;
        transition:0.3s;
    }}

    .card:hover {{
        transform:translateY(-5px);
    }}

    footer {{
        padding:20px;
        text-align:center;
        background:var(--surface);
    }}
    """

    return css
```

```python
class HTMLPageBuilder:

    def hero_section(self, title):

        return f"""
        <section class="hero">
        <h1>{title}</h1>
        <p>AI Generated Website</p>
        <button>Get Started</button>
        </section>
        """

    def features_section(self):

        return """
        <section class="features">

        <div class="card">
        <h3>Fast</h3>
        <p>Generate websites instantly</p>
        </div>

        <div class="card">
        <h3>Responsive</h3>
        <p>Works on all devices</p>
        </div>

        <div class="card">
        <h3>Modern</h3>
        <p>Beautiful UI Design</p>
        </div>

        </section>
        """

    def contact_section(self):

        return """
        <section style="padding:40px;text-align:center">

        <h2>Contact Us</h2>

        <input placeholder="Name"><br><br>
        <input placeholder="Email"><br><br>
        <textarea placeholder="Message"></textarea><br><br>

        <button>Send</button>

        </section>
        """

    def footer(self):

        return """
        <footer>
        <p>© 2026 HTML Quick Styler</p>
        </footer>
        """

    def build_page(self, title, palette):

        css = generate_css(palette)

        html = f"""
        <!DOCTYPE html>
        <html>

        <head>
        <title>{title}</title>

        <style>
        {css}
        </style>

        </head>

        <body>

        {self.hero_section(title)}

        {self.features_section()}

        {self.contact_section()}

        {self.footer()}

        </body>
        </html>
        """

        return html
```

```python
palette_generator = ColorPaletteGenerator()

page_builder = HTMLPageBuilder()
```

```python
def generate_page(title, style):

    palette = palette_generator.get_palette(style)

    html = page_builder.build_page(title, palette)

    return html, html
```

```python
with gr.Blocks() as app:

    gr.Markdown("# HTML Quick Styler")

    with gr.Tab("Generate Page"):

        title = gr.Textbox(label="Website Title")

        style = gr.Dropdown(
            ["modern","ocean","forest","sunset"],
            label="Color Style"
        )

        generate_btn = gr.Button("Generate Website")

        html_code = gr.Code(language="html")

        preview = gr.HTML()

        generate_btn.click(
            generate_page,
            inputs=[title,style],
            outputs=[html_code,preview]
        )

app.launch(share=True)
```
